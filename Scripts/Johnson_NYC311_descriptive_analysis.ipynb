```python
#import packages
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import urllib.request, json
import datetime
```

### Using socrata API we can access NYC 311 data going back to 2004:

NYC API url: https://data.cityofnewyork.us/resource/erm2-nwe9.json

```python
#query 2022 data and store in df
url = 'https://data.cityofnewyork.us/resource/erm2-nwe9.json'

query = (url+'?'
        "$select=*"
        "&$where=%s between '2022-01-01T00:00:00.000' and '2022-12-31T23:59:59.000'")%('created_date')
query = query.replace(" ", "%20")
response = urllib.request.urlopen(query)
data = json.loads(response.read())

#store in a dataframe
df = pd.DataFrame(data, columns = data[0].keys())
df = df.drop([cols for cols in data[0].keys() if cols.find("@computed_region_") != -1], axis = 1)

#convert df time columns to datetime data type
df['created_date'] = pd.to_datetime(df['created_date'])
```
